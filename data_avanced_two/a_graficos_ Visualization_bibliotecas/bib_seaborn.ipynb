```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set_theme() #setando o tema padrão
```

```python
df = pd.read_csv('./datas/canadian_immegration_data.csv')
```

```python
#Vamos realizar uma analise que sera uma visualização global da imigração para o Canada um top 10
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
top_10
```

```python
sns.barplot(data=top_10, x=top_10.Country, y='Total')
```

```python
# This is the modern, future-proof way to get the same result
sns.barplot(data=top_10, y='Country', x='Total', hue='Country', legend=False, palette='viridis')
```

```python
ax = sns.barplot(data=top_10,  y='Country', x='Total', orient='h',hue='Country',legend=False,palette='viridis')

ax.set(title="Países com maior imigração para o Canadá\n1980 a 2013",
       xlabel='Número de Imigrantes',
        ylabel='')

plt.show()
```

```python
fig ,ax = (plt.subplots(figsize=(8,4)))
#posso atribuir paletas mais acertivas dessa forma aqui 
# # Passo 1: Use a "fábrica" para criar sua lista de 10 cores (pois temos 10 países).
minha_paleta_customizada = sns.color_palette("husl", 10) 
ax = sns.barplot(data=top_10,  y='Country', x='Total', orient='h',hue='Country',legend=False,palette=minha_paleta_customizada)
ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013',loc='left',fontsize=16)
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
```

```python
# Podemos sempre Otimizar isso tudo coloando em uma função 
def gerar_grafico(minha_paleta_customizada):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax = sns.barplot(data=top_10,  y='Country', x='Total', orient='h',hue='Country',legend=False,palette=minha_paleta_customizada)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013',loc='left',fontsize=16)
    ax.set_xlabel('Número de Imigrantes',fontsize=14)
    ax.set_ylabel('')
    sns.despine() # Retira as linhas de cima e do lado
    plt.show()
# Podemos  tambem por o dataframe em uma variavel e chama-lo na função tambem , pode ser muito util no futuro
```

```python
gerar_grafico(sns.color_palette("husl", 10))
# ou
gerar_grafico("Blues_r") # o _r inverte as cores deixando o mais forte para os graficos com numeros maiores 
# ou
gerar_grafico("rocket")
# ou
gerar_grafico("coolwarm") # Quente e frio
# ou
gerar_grafico("tab10") # Palleta categorica , seria perfeita para o nosso caso que temos cada País
```

```python
sns.set_theme(style='white') #Aqui com apenas o white retiramos as grids se quisermos por a grid novamente só adicionar a palavra grid
gerar_grafico("rocket")
```

```python
sns.set_theme(style='ticks') # Com tracinhos 
gerar_grafico("rocket")
```
